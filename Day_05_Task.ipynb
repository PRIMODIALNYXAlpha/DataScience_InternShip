# 3) Rectangle class with area and perimeter

```python
print("3) Rectangle Class")

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

r1 = Rectangle(10, 5)
print("Area:", r1.area())
print("Perimeter:", r1.perimeter())

print("\n" + "-"*60 + "\n")
```

# 6) BankAccount with private balance
# deposit, withdraw, get_balance

```python
print("6) BankAccount Class (Private Balance)")

class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return "Deposit Successful ‚úÖ"
        return "Invalid deposit amount ‚ùå"

    def withdraw(self, amount):
        if amount <= 0:
            return "Invalid withdrawal amount ‚ùå"
        if amount > self.__balance:
            return "Insufficient balance ‚ùå"
        self.__balance -= amount
        return "Withdrawal Successful ‚úÖ"

    def get_balance(self):
        return self.__balance

acc = BankAccount(5000)
print(acc.deposit(2000))
print(acc.withdraw(1000))
print("Balance:", acc.get_balance())

print("\n" + "-"*60 + "\n")
```

# 7) User class: private password
# allow password update using setter method

```python
print("7) User Class (Private Password + Setter)")

class User:
    def __init__(self, username, password):
        self.username = username
        self.__password = password

    def set_password(self, new_password):
        if len(new_password) < 6:
            return "Password too short ‚ùå"
        self.__password = new_password
        return "Password updated ‚úÖ"

    def check_password(self, password):
        return password == self.__password

u1 = User("tarun", "OldPass123")
print(u1.check_password("OldPass123"))
print(u1.set_password("NewPass456"))
print(u1.check_password("NewPass456"))

print("\n" + "-"*60 + "\n")
```

# 9) ATM class: hide pin, allow transaction after validation

```python
print("9) ATM Class (Hidden PIN)")

class ATM:
    def __init__(self, pin, balance=0):
        self.__pin = pin
        self.__balance = balance

    def validate_pin(self, entered_pin):
        return entered_pin == self.__pin

    def withdraw(self, entered_pin, amount):
        if not self.validate_pin(entered_pin):
            return "Invalid PIN ‚ùå"
        if amount > self.__balance:
            return "Insufficient balance ‚ùå"
        self.__balance -= amount
        return f"Withdraw successful ‚úÖ Remaining balance: {self.__balance}"

atm = ATM(1234, 8000)
print(atm.withdraw(1234, 2000))
print(atm.withdraw(1111, 1000))

print("\n" + "-"*60 + "\n")
```

# 14) Shape -> Circle and Rectangle

```python
print("14) Shape Inheritance")

import math

class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

class Rectangle2(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

print("Circle area:", Circle(5).area())
print("Rectangle area:", Rectangle2(10, 4).area())

print("\n" + "-"*60 + "\n")
```

# 16) Override calculate_salary() in Manager and Intern

```python
print("16) Method Overriding - calculate_salary()")

class EmployeeBase:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def calculate_salary(self):
        return self.base_salary

class Manager(EmployeeBase):
    def calculate_salary(self):
        return self.base_salary + 20000  # bonus

class Intern(EmployeeBase):
    def calculate_salary(self):
        return self.base_salary - 5000   # stipend cut

m = Manager("Tarun", 50000)
i = Intern("Rahul", 20000)

print("Manager Salary:", m.calculate_salary())
print("Intern Salary:", i.calculate_salary())

print("\n" + "-"*60 + "\n")
```

# MINI PROJECT 1: Library Management System
# using class, inheritance, encapsulation

```python
print("MINI PROJECT 1: Library Management System\n")

class LibraryItem:
    def __init__(self, title):
        self.title = title

class LibraryBook(LibraryItem):
    def __init__(self, title, author):
        super().__init__(title)
        self.author = author
        self.__is_issued = False  # encapsulation

    def issue(self):
        if self.__is_issued:
            return "Book already issued ‚ùå"
        self.__is_issued = True
        return "Book issued successfully ‚úÖ"

    def return_book(self):
        if not self.__is_issued:
            return "Book was not issued ‚ùå"
        self.__is_issued = False
        return "Book returned successfully ‚úÖ"

    def status(self):
        return "Issued" if self.__is_issued else "Available"

    def display(self):
        print(f"Title: {self.title}, Author: {self.author}, Status: {self.status()}")

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def show_books(self):
        for b in self.books:
            b.display()

lib = Library()
lib.add_book(LibraryBook("Ikigai", "Hector Garcia"))
lib.add_book(LibraryBook("Atomic Habits", "James Clear"))

lib.show_books()
print(lib.books[0].issue())
print(lib.books[0].issue())
print(lib.books[0].return_book())

print("\n" + "="*70 + "\n")
```

# MINI PROJECT 2: Student Result System with grading logic

```python
print("MINI PROJECT 2: Student Result System\n")

class StudentResult:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def grade(self):
        avg = sum(self.marks) / len(self.marks)
        if avg >= 80:
            return "A"
        elif avg >= 60:
            return "B"
        elif avg >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}, Grade: {self.grade()}")

sr1 = StudentResult("Tarun", [80, 75, 88, 92])
sr1.display()

print("\n" + "="*70 + "\n")
```

# ONLINE SHOPPING SYSTEM (FULL)
# Product base class
# Electronics / Clothing subclasses
# Cart class
# Payment method
# Bill generation

```python
print("ONLINE SHOPPING SYSTEM\n")

class ProductBase:
    def __init__(self, name, price):
        self.name = name
        self._price = price

    def get_price(self):
        return self._price

    def display(self):
        print(f"{self.name} - ‚Çπ{self._price}")

class Electronics2(ProductBase):
    def __init__(self, name, price, warranty):
        super().__init__(name, price)
        self.warranty = warranty

    def display(self):
        print(f"Electronics: {self.name} - ‚Çπ{self._price} | Warranty: {self.warranty} years")

class Clothing2(ProductBase):
    def __init__(self, name, price, size):
        super().__init__(name, price)
        self.size = size

    def display(self):
        print(f"Clothing: {self.name} - ‚Çπ{self._price} | Size: {self.size}")

class Cart:
    def __init__(self):
        self.items = []

    def add_item(self, product):
        self.items.append(product)

    def total(self):
        return sum(item.get_price() for item in self.items)

    def show_cart(self):
        print("CART ITEMS:")
        for item in self.items:
            item.display()

class PaymentMethod:
    def pay(self, amount):
        pass

class UPI2(PaymentMethod):
    def pay(self, amount):
        return f"Payment of ‚Çπ{amount} done via UPI ‚úÖ"

class Cash2(PaymentMethod):
    def pay(self, amount):
        return f"Payment of ‚Çπ{amount} done via Cash ‚úÖ"

def generate_bill(cart, payment_method):
    cart.show_cart()
    total_amount = cart.total()
    print("\nTotal Amount:", total_amount)
    print(payment_method.pay(total_amount))
    print("\nüßæ BILL GENERATED SUCCESSFULLY!")

# Demo
cart = Cart()
cart.add_item(Electronics2("Mobile", 25000, 1))
cart.add_item(Clothing2("Jeans", 1500, "32"))

generate_bill(cart, UPI2())

print("\n‚úÖ DAY 5 COMPLETED SUCCESSFULLY!")
```
